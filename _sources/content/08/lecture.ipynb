## Identifying Housing and Homeless Meetings in Seattle

* import and pull down data, create chunks, embed, and cluster, UMAP, cluster, and inspect clusters

* we will use a new embedding model with a larger context length to allow for as much of the meeting transcript to be embedded at once
    * https://huggingface.co/jinaai/jina-embeddings-v2-base-en

```python
from cdp_data import CDPInstances, datasets

sessions = datasets.get_session_dataset(
    CDPInstances.Seattle,
    start_datetime="2021-01-01",
    end_datetime="2022-01-01",
    store_transcript=True,
    raise_on_error=False,
)
sessions
```

```python
# load sentence transformers to embed
# chunks of meetings at a time and
# store in a dict with the session id
# and a chunk index as the key
from cdp_backend.pipeline.transcript_model import Transcript
from tqdm import tqdm

# we are going to use a new chunking strategy
# instead of chunking by a fixed number of sentences
# we are going to chunk by a target number of words, to the nearest sentence
# this is because we want to keep the context of the sentence
# and not split sentences in half
target_n_words = 3000

session_chunk_id_to_chunk_text = {}
for _, row in tqdm(
    sessions.iterrows(),
    desc="Processing sessions",
    total=len(sessions),
):
    # Read transcript
    with open(row.transcript_path) as open_f:
        transcript = Transcript.from_json(open_f.read())

    # Iter sentences
    chunk_i = 0
    current_chunk_sentences = []
    current_chunk_n_words = 0
    for sentence in transcript.sentences:
        next_addition_n_words = len(sentence.words)
        
        # if the current chunk length + the next sentence length
        # is greater than the target length, then embed the chunk
        # and reset the chunk
        if current_chunk_n_words + next_addition_n_words > target_n_words:
            # embed the chunk
            session_chunk_id_to_chunk_text[
                f"{row['id']}-{chunk_i}"
            ] = " ".join(current_chunk_sentences)

            # reset the chunk
            chunk_i += 1
            current_chunk_sentences = []
            current_chunk_n_words = 0

        # add the sentence to the chunk
        current_chunk_sentences.append(sentence.text)
        current_chunk_n_words += next_addition_n_words
```

* Just for debugging, lets quickly print some stats to ensure that we are getting the right size chunks

```python
lengths_of_chunks = [len(v.split()) for v in session_chunk_id_to_chunk_text.values()]
print(f"Average chunk length: {sum(lengths_of_chunks) / len(lengths_of_chunks)}")
print(f"Minimum chunk length: {min(lengths_of_chunks)}")
```

```python
# checkpoint save to parquet
import pandas as pd
# Convert the chunk dict into a dataframe
session_chunks = pd.DataFrame(
    [
        {"session_chunk_id": k, "text": v}
        for k, v in session_chunk_id_to_chunk_text.items()
    ]
)

# Save to parquet
session_chunks.to_parquet("session_chunks.parquet")
```

* now we can embed all of these chunks using our new model

```python
from transformers import AutoModel
import pandas as pd

# Load the model
# this is what sentence-transformers does for us
# but we need to do it manually here so that we can trust the model
model = AutoModel.from_pretrained(
    "jinaai/jina-embeddings-v2-base-en",
    trust_remote_code=True,
)

# Load the chunks
session_chunks = pd.read_parquet("session_chunks.parquet")

# Embed the chunks
session_chunks["embedding"] = model.encode(
    session_chunks["text"].tolist(),
    batch_size=1,  # this is variable and depends on your machine
    show_progress_bar=True,
)

session_chunks
```
